### Задание 1
1. Импортируйте библиотеки pandas, numpy и matplotlib.
2. Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
data = boston["data"]
feature_names = boston["feature_names"]
target = boston["target"]
```

3. Создайте датафреймы X и y из этих данных.

```python
X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(target, columns=["price"])
```

4. Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test)
с помощью функции train_test_split так, чтобы размер тестовой выборки
составлял 20% от всех данных, при этом аргумент random_state должен быть равен 42.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

5. Масштабируйте данные с помощью StandardScaler.

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=feature_names)

X_test_scaled = scaler.fit_transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=feature_names)
```

6. Постройте модель TSNE на тренировочный данных с параметрами:
n_components=2, learning_rate=250, random_state=42.


```python
model_tsne = TSNE(n_components=2, learning_rate=250, random_state=42)
model = model_tsne.fit_transform(X_train_scaled)
```

7. Постройте диаграмму рассеяния на этих данных.

```python
plt.style.use('classic') #Тут я доигрался со стилями и у меня пропали подписи к х и у, хотя скорее всего они просто белые
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
plt.scatter(model[:, 0], model[:, 1])
plt.show()
```

### Задание 2

1. С помощью KMeans разбейте данные из тренировочного набора на 3 кластера,
используйте все признаки из датафрейма X_train.
Параметр max_iter должен быть равен 100, random_state сделайте равным 42.



```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters=3, max_iter=100, random_state=42)
kmeans = kmeans_model.fit_predict(X_train)
```

2. Постройте еще раз диаграмму рассеяния на данных, полученных с помощью TSNE,
и раскрасьте точки из разных кластеров разными цветами.


```python
plt.scatter(model[:, 0], model[:, 1], c=kmeans) # берем срезы и в качестве цвета (с) - ставим значения kmeans
plt.show()
```

3. Вычислите средние значения price и CRIM в разных кластерах.

```python
print ('*'*15 + 'CLASTERS\' AVERAGE PRICE' + '*'*15)

for x in range(3):
    print(f'Claster\'s {x} average price is {y_train[kmeans == x].mean().values.round(2)}')

    
print ('\n' + '*'*16 + 'CLASTERS\' CRIM SCORE' + '*'*17)
for x in range(3):
    a = X_train.loc[kmeans == x, 'CRIM'].mean()
    print(f'Claster\'s {x} CRIM score is {np.round(a,2)}')
```

2. Вычислите средние значения price и CRIM в разных кластерах на тестовых данных.
 

```python
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
X_test_tsne = model_tsne.fit_transform(X_test_scaled)

tries = kmeans_model.fit_predict(X_test_scaled)

plt.scatter(X_test_tsne[:, 0], X_test_tsne[:, 1], c=tries)
plt.show()
```
