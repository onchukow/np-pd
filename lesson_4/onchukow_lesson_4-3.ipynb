```python
"""
Можно ли массив Numpy, состоящий из одного столбца и 12 строк, назвать одномерным?
"""
# Для проверки этого условия создадим вектор Х 
x = np.arange(1,13)

# На втором этапе мы развернем его по вертикали
y = x.reshape(12,1)

# На третьем развернем обратно по горизонтали
z = y.reshape(1,12)
x,y,z

print('Вектор:', x)
print("Первое колдовство:", y)
print("Попытка вернуть как было:", z)

print('*' * 50)

# Что мы видим - если на начальном этапе массив был одномерным, то после колдовства - он превратился в двумерный 
# (еще на первом этапе магии). 
# А именно двумерным массивом с одним столбцом и строками внутри. Это видно по дополнительным []. 
# Также это можно проверить индексами, которые становятся двумерными и требуют двумерный индекс

print ("Векторный индекс [0]:", x[0])
print ("Тот же индекс два раза перевернутого вектора [0]:", z[0])
print ("Использование двойных индексов [0][0], чтобы получить значение из п.1:", z[0][0])
```

```python
"""Вертикально соединить массивы a и b. a - двумерный массив из нулей, число строк которого 
больше 1 и на 1 меньше, чем число строк двумерного массива b, состоящего из единиц. 
Итоговый массив v должен иметь атрибут size, равный 10.
"""

a = np.zeros((2,2))

print('Zeros a: \n', a)

b = np.ones((3,2))
print('Ones b: \n', b)

v = np.vstack((a, b))
print('Collab v: \n',v)

print('\n', '*' * 50)
print('\nV size: ',v.size)
```

```python
"""
1. Создайте одномерный массив c, составленный из последовательности 16-ти случайных 
равномерно распределенных целых чисел от 0 до 16 невключительно. 

2. Поменяйте его форму так, чтобы получилась квадратная матрица C. 

3. Получите матрицу D, поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10. 

4. Вычислите определитель, ранг и обратную матрицу D_inv для D.

"""
from math import sqrt

c = np.random.randint(0, 16,16)

c_calc = int(sqrt(c.size)) # Автоматическое создание квадратных матриц (если это возможно и при идеальном пользователе)

C = c.reshape(c_calc,c_calc)

D = (b + (np.dot(C,10))) # b взята из предыдущего задания. Матрица С умножается на скаляр

d_opr = np.linalg.det(D)
d_rang = np.linalg.matrix_rank(D)
D_inv = np.linalg.inv(D)

print(f'Определитель D: {d_opr}\nРанг D: {d_rang}\nОбратная матрица D: \n{D_inv}\n')
```
